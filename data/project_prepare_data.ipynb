```python
from pathlib import Path

def merge_JsonFiles(folder, file_name):
    filepath = Path('/content/'+file_name+".csv")
    filepath.parent.mkdir(parents=True, exist_ok=True)
    result = list()
    for f1 in folder:
      file1 = extract_json(f1)
      file1.open_file()
      coredata = file1.coredata()
      author = file1.authors()
      aff = file1.affiliation()
      subject =file1.subject_areas()
      keywords =file1.authkeywords()
      mainterm = file1.idxterm()
      language = file1.language()
      item = file1.item()
      main_dict =dict()
      main_dict.update(author)
      main_dict.update(aff)
      main_dict.update(coredata)
      main_dict.update(keywords)
      main_dict.update(mainterm)
      main_dict.update(subject)
      main_dict.update(language)
      main_dict.update(item)
      result.append((main_dict))
    df = pd.DataFrame(result)
    df.to_csv(filepath)
    return df
```
