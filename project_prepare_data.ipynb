```python

class extract_json:
  def __init__(self, file_path):
    self.file = file_path
  def open_file(self):
    with open(self.file) as fd:
      self.cat = json.load(fd)
    print(self.file)
  def bibliogrphy(self):
    tail = self.cat['abstracts-retrieval-response']['item']['bibrecord']['tail']
    #print(bool(tail))
    if not bool(tail) or not bool(tail['bibliography']["reference"]["ref-info"]["ref-title"]["ref-titletext"]):
      self.num_ref = 0
      self.ref_title = "null"
    else:
      self.bib = tail['bibliography']
      self.num_ref = int(self.bib["@refcount"])
    #print(num_ref)
      self.ref_title =[]
      if self.num_ref ==1:
        title=self.bib["reference"]["ref-info"]["ref-title"]["ref-titletext"]
        self.ref_title.append(title)
      else:
        for i in range(self.num_ref):
          if "ref-title" in self.bib["reference"][i]["ref-info"]:
            title = self.bib["reference"][i]["ref-info"]["ref-title"]["ref-titletext"]
          else:
            title = "null"
          self.ref_title.append(title)
      self.ref_title = self.ref_title
    return (self.num_ref, self.ref_title)
  def coredata(self):
    self.core = self.cat['abstracts-retrieval-response']['coredata']
    self.title = self.core["dc:title"]
    self.pbname = self.core["prism:publicationName"]
    self.source_name = self.core["dc:publisher"]
    self.authurs = self.core["dc:creator"]["author"]
    self.cite = int(self.core["citedby-count"])

    return (self.title, self.pbname, self.source_name, self.authurs, self.cite )
  def get_abstract(self):
    #print(bool(self.cat['abstracts-retrieval-response']['item']['bibrecord']["head"]["abstracts"]))
    if bool(self.cat['abstracts-retrieval-response']['item']['bibrecord']["head"]["abstracts"]):
            self.abstract = self.cat['abstracts-retrieval-response']['item']['bibrecord']["head"]["abstracts"]
    else:
            self.abstract = "null"
    return self.abstract
  def get_title(self):
    a = "dc:title"
    if bool(self.cat['abstracts-retrieval-response']['coredata']) and a in self.cat['abstracts-retrieval-response']['coredata']:
      self.title = self.cat['abstracts-retrieval-response']['coredata']["dc:title"]
    else:
      self.title = "null"
    return self.title
  def get_keyword(self):
    m ="mainterm"
    if bool(self.cat['abstracts-retrieval-response']["idxterms"]) and m in self.cat['abstracts-retrieval-response']["idxterms"]:
      self.mainterm =self.cat['abstracts-retrieval-response']["idxterms"]["mainterm"]
      k_list =list()
      #print(type(self.mainterm))
      if type(self.mainterm)==list:
        for i in self.mainterm:
          k_list.append(i["$"])
      else:
        k_list.append(self.mainterm["$"])
      self.keyword = k_list
    else:
      self.keyword = "null"
    return self.keyword



```

```python
from pathlib import Path
def merge_JsonFiles(folder, file_name):
    filepath = Path('/content/'+file_name+".csv")
    filepath.parent.mkdir(parents=True, exist_ok=True)
    result = list()
    for f1 in folder:
        file1 = extract_json(f1)
        file1.open_file()
        """
        num_ref,ref_title = file1.bibliogrphy()

        title,pbname, source_name, authors, cite = file1.coredata()
        """
        abstract = file1.get_abstract()
        title = file1.get_title()
        keyword = file1.get_keyword()
        result.append((title,abstract,keyword))
    df = pd.DataFrame(result, columns = ["title","Absract","Keyword"])
    df.to_csv(filepath)
    return result
```
